# Extracting Useful Info From a Website Using Beautifulsoup

```python
import urllib
import bs4 as bs
import nltk
from nltk.corpus import stopwords
```

```python
source=urllib.request.urlopen("https://en.wikipedia.org/wiki/Water").read()
```

```python
soup=bs.BeautifulSoup(source,"lxml")
```

```python
soup
```

```python
text=""
for text1 in soup.find_all("p"):
    text+=text1.text
```

```python
text=text.lower()
```

```python
import re
text=re.sub(r"\[[0-9]*\]"," ",text)
```

```python
text=re.sub(r"\n"," ",text)
```
